```
x = torch.rand(2,2)
y = torch.rand(2,2)

print(torch.add(x,y)) # 일반적인 사칙연산
print(x+y)
# torch.mul
# torch.div
# ...
```

```
print(y)
y.add_(x) # _ (underbar) 를 통해 inplace 역할
print(y)
```

```
x = torch.rand(5,5)
print(x)
print(x[:,0]) # 원하는 row, col
```

```
# reshape

x = torch.rand(2,2)
print(x)

y = x.view(4)
print(y)

z = x.view(-1, 4)
print(z)
```

```
# convert numpy to torch.tensor
import numpy as np

a = torch.ones(5)
print(a)

b = a.numpy() # tensor to numpy
print(b)
print(type(b))

a.add_(1)
print(a)
print(b) # b도 변해버린다
```

```
a = np.ones(4)
print(a)
b = torch.from_numpy(a)
print(b)

a += 1
print(a)
print(b) # 여기도 마찬가지
```

```
if torch.cuda.is_available():
    device = torch.device("cuda") # GPU 사용 가능 확인

x = torch.ones(4, device=device) # 처음부터 GPU로 만들기

y = torch.ones(5)
y = y.to(device) # GPU로 옮기기

z = x + y # GPU에서 계산

z = z.to("cpu") # 다시 CPU로 계산
```

```
# requires_grad
# 해당 tensor에서 grad 저장
# 다음 tutorial에서 배운다

x = torch.ones(5, requires_grad=True)
print(x)
```
